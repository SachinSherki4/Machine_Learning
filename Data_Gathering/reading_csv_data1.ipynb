### converting All teams full names to short form for easy reading

```python
# writing function to convert full form to short forms
def team_short_forms(team):
    if team == "Royal Challengers Bangalore":
        return 'RCB'
    elif team == "Kolkata Knight Riders":
        return 'KKR'
    elif team == "Mumbai Indians":
        return 'MI'
    elif team == "Gujarat Lions":
        return 'GL'
    elif team == "Rising Pune Supergiant":
        return 'RPS'
    elif team == "Delhi Daredevils":
        return 'DD'
    elif team == "Kings XI Punjab":
        return 'KP'
    else :
        return team
```

```python
ipl1=pd.read_csv('../../Data_Files/IP_Data_Sets/matches.csv', converters={'team1': team_short_forms })
ipl1.head()
```

### loading only dat with population more than 400 and less that 450

```python
housing_filter_data =[]

for chunks in pd.read_csv('../../Data_Files/Lesson_03_Supervised_Learning_ Regression_and_its_Application/housing_with_ocean_proximity.csv', chunksize=5500):
    filtered= chunks[(chunks['population'] > 400) & (chunks['population'] < 450)]
    housing_filter_data.append(filtered)

final_df= pd.concat(housing_filter_data)
print(final_df.info())
```

### Calculating mean without loading data fully

```python
house_value=0
house_count=0
for housing_data in pd.read_csv('../../Data_Files/Lesson_03_Supervised_Learning_ Regression_and_its_Application/housing_with_ocean_proximity.csv', chunksize=5500):
    house_value += housing_data['median_house_value'].sum()
    house_count += housing_data['median_house_value'].count()

avg_house_value=house_value/house_count
avg_house_value
```

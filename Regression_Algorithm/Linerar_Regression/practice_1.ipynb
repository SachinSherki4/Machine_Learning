```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
adv_df= pd.read_csv('../../Data_Files/Supervise_learning/advertising.csv')
```

```python
adv_df.info()
```

```python
adv_df.head()
```

```python
plt.Figure(figsize=(8,6))
plt.scatter(adv_df['TV'], adv_df['Sales'])
plt.title('TV vs Sales')
```

```python
sm_tv_sales_model=smf.ols(formula='Sales ~ TV', data=adv_df).fit()
```

```python
print(sm_tv_sales_model.summary())
```

```python
plt.Figure(figsize=(8,6))
sns.scatterplot(x='Radio',y='Sales', data=adv_df)
plt.title('Radio vs Sales')
```

```python
sm_radio_sales_model= smf.ols(formula='Sales ~ Radio', data=adv_df).fit()
```

```python
print(sm_radio_sales_model.summary())
```

```python
plt.Figure(figsize=(8,6))
plt.scatter(adv_df['Newspaper'], adv_df['Sales'])
plt.title('Newspaper vs Sales')
```

```python
sm_newspaper_sales_model=smf.ols(formula='Sales ~ Newspaper', data=adv_df).fit()
```

```python
print(sm_newspaper_sales_model.summary())
```

```python
sm_tv_radio_newspaper_model=smf.ols(formula='Sales ~ TV + Radio + Newspaper', data=adv_df).fit()
```

```python
print(sm_tv_radio_newspaper_model.summary())
```
